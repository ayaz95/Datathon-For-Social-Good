# Reverse Geocoding to get addresses corresponding to Hubs

```python
# Importing all the required libraries
import geopandas as gpd
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import plotly_express as px
import tqdm
from tqdm.notebook import tqdm_notebook
```

```python
# Geocoder used : Open Street Map
locator = Nominatim(user_agent="https://nominatim.openstreetmap.org/")
```

```python
# Checked if geocoder is working properely by taking one record
coordinates = '-33.9068, 151.05015'
location = locator.reverse(coordinates, timeout=10)
# The address info fetched by geocoder for coordinates in raw form
location.raw
```

```python
# Reading Hub coordinates file
path= 'D:\Jupyter_Notebooks\Hub Locations.xlsx'
xl = pd.ExcelFile(path)
print(xl.sheet_names)
```

```python
# Sheet 2
hub1 = xl.parse('Sheet2')
```

```python
# First 5 records display
hub1.head()
```

```python
# Number of rows and columns in the data
hub1.shape
```

```python
# Removing all records with no HubID
hub1 = hub1[hub1['Hub ID'].notna()]
```

```python
# Number of records after removing NAs or irrelevant information for analysis
hub1.shape
```

```python
# Splitting the coordinated into lat and lon, addition of two new columns
hub1[['lat','lon']] = hub1["Location Coordinates"].str.split("|",expand=True,)
```

```python
# displaying first 5 records
hub1.head()
```

```python
# Understanding data types
hub1.dtypes
```

```python
# Creating geometry as a column to be readeable by nominatim( geocoder)
hub1["geometry"] = hub1["lat"] + ',' + hub1["lon"]
```

```python
hub1['geometry'][0]
hub1.head()
```

```python
locator = Nominatim(user_agent="https://nominatim.openstreetmap.org/", timeout=10)
rgeocode = RateLimiter(locator.reverse, min_delay_seconds=0.001)
```

```python
tqdm_notebook().pandas()
```

```python
# Getting addresses corresponding to coordinates
hub1['address'] = hub1['geometry'].progress_apply(rgeocode)
```

```python
#Displaying first 5 records
hub1.head()
```

```python
# Displaying all the records of addresses, total =70
print(hub1['address'])
```

```python
#Address column consists of addresses as well as coordinates, separating columns in tow parts
df=pd.DataFrame((hub1['address'].values.tolist()),hub1['Hub ID'])
df.head()
```

```python
# Splitting address column in number of columns by separator ','
df = df[0].apply(lambda x: pd.Series(x.split(',')))
# Displaying first 5 records
df.head(20)
```

```python
# There can be better logic to extract post codes from address using address parsing libaray 
# but as data is small and understandable
# Hence, observed column 3, 4, 5, 6 contains postal codes
# dropping all unnecassary columns
df= df.drop([0,1,2,7,8], axis=1)
# Replacing value as 0 for all non numeric vales, easy to fetch postal codes
cols = df.columns
df= df[cols].apply(pd.to_numeric, errors='coerce').fillna(0)
# adding postal code column
df['post code']= df[3]+df[4]+df[5]+df[6]
# dropping all columns, keeping only postal code and hub id
df= df.drop([3,4,5,6], axis=1)
df.head()
```

```python
# Exporting results in excel file by creating a new file in the directory
df.to_excel(r'D:\Jupyter_Notebooks\Hubswith_postalCodes.xlsx')
```
