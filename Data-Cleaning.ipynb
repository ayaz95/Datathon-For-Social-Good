```python
#Calculating the number of females between age 0-4 for all the states



###### NSW
no_of_females = []

for k in df['Code']:
    if str(k).startswith('1'):
        no_of_females.append(df.loc[df['Code'] == k, 'Number_0-4'].iloc[0])
        
no_of_females_0_4_NSW = sum(no_of_females)
print("New South Wales Females(0-4) : ", round(no_of_females_0_4_NSW))



###### Victoria
no_of_females = []

for k in df['Code']:
    if str(k).startswith('2'):
        no_of_females.append(df.loc[df['Code'] == k, 'Number_0-4'].iloc[0])
        
no_of_females_0_4_VIC = sum(no_of_females)
print("Victoria State Females(0-4) : ", round(no_of_females_0_4_VIC))


###### Queensland
no_of_females = []

for k in df['Code']:
    if str(k).startswith('3'):
        no_of_females.append(df.loc[df['Code'] == k, 'Number_0-4'].iloc[0])
        
no_of_females_0_4_Queensland = sum(no_of_females)
print("Queensland Females(0-4) : ",round(no_of_females_0_4_Queensland))


###### South Australia
no_of_females = []

for k in df['Code']:
    if str(k).startswith('4'):
        no_of_females.append(df.loc[df['Code'] == k, 'Number_0-4'].iloc[0])
        
no_of_females_0_4_SA = sum(no_of_females)
print("South Australia Females(0-4) : ", round(no_of_females_0_4_SA))



#### Western Australia

no_of_females = []

for k in df['Code']:
    if str(k).startswith('5'):
        no_of_females.append(df.loc[df['Code'] == k, 'Number_0-4'].iloc[0])
        
no_of_females_0_4_WA = sum(no_of_females)
print("Western Australia Females(0-4) : ",round(no_of_females_0_4_WA))


#### Tasmania

no_of_females = []

for k in df['Code']:
    if str(k).startswith('6'):
        no_of_females.append(df.loc[df['Code'] == k, 'Number_0-4'].iloc[0])
        
no_of_females_0_4_TAS = sum(no_of_females)
print("Tasmania Females(0-4) : ",round(no_of_females_0_4_TAS))


#### Northern Territory
no_of_females = []

for k in df['Code']:
    if str(k).startswith('7'):
        no_of_females.append(df.loc[df['Code'] == k, 'Number_0-4'].iloc[0])
        
no_of_females_0_4_NT = sum(no_of_females)
print("Northern Territory Females(0-4) : ",round(no_of_females_0_4_NT))
```
